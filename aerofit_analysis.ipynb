> ## **Outlier Detection**


```python
for column in ['Age', 'Education', 'Usage', 'Fitness', 'Income', 'Miles']:
    print(f"Non-numeric values in {column}:")
    print(data[column][pd.to_numeric(data[column], errors='coerce').isna()])
    print("\n")
```

```python
for column in ['Age', 'Education', 'Usage', 'Fitness', 'Income', 'Miles']:
    data[column] = pd.to_numeric(data[column], errors='coerce')

Q1 = data[['Age', 'Education', 'Usage', 'Fitness', 'Income', 'Miles']].quantile(0.25)
Q3 = data[['Age', 'Education', 'Usage', 'Fitness', 'Income', 'Miles']].quantile(0.75)

IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = ((data[['Age', 'Education', 'Usage', 'Fitness', 'Income', 'Miles']] < lower_bound) |
            (data[['Age', 'Education', 'Usage', 'Fitness', 'Income', 'Miles']] > upper_bound))

print(outliers.sum())
sns.boxplot(x=data['Income'])
plt.show()
```

```python
import pandas as pd
product_distribution = data['Product'].value_counts(normalize=True) * 100
print(product_distribution)

plt.figure(figsize=(8, 6))
sns.barplot(x=product_distribution.index, y=product_distribution.values, color=sns.color_palette('viridis')[0])
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.title('Percentage Distribution of Products', fontsize=16)
plt.xlabel('Product', fontsize=14)
plt.ylabel('Percentage (%)', fontsize=14)
plt.show()
print(product_distribution)
```

* **Product – Gender**

```python
male_percentage = data[data['Gender'] == 'Male']['Product'].count() / len(data) * 100
print("Percentage of Male customers purchasing treadmill: ", male_percentage)

female_percentage = len(data[(data['Gender'] == 'Female') & (data['Product'] == 'KP281')]) / len(data) * 100
print("Percentage of Female customers purchasing KP281 treadmill: ", female_percentage)
```

```python
percentages = [male_percentage, female_percentage]
categories = ['Male', 'Female (KP281)']

plt.figure(figsize=(8, 6))

sns.barplot(x=categories, y=percentages, palette='Set2')

plt.title('Percentage Distribution of Customers Purchasing Treadmills', fontsize=16)
plt.xlabel('Customer Category', fontsize=14)
plt.ylabel('Percentage (%)', fontsize=14)
plt.xticks(rotation=0, fontsize=12)
plt.yticks(fontsize=12)
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
